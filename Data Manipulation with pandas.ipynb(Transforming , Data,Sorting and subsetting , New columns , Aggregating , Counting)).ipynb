# Aggregating Data

```python
shopping_trends.head()
```

```python
# STATISTICS
print(f"Media: {shopping_trends['Purchase Amount (USD)'].mean():.2f}")
print(f"Max: {shopping_trends['Purchase Amount (USD)'].max()}")

# Other statistics like
# .median() || .mode() || .min() || .max() ||
#      .var() || .std() || .sum() || .quantile()

def pct40(column):
    return column.quantile(0.4)

def pct30(column):
    return column.quantile(0.3)


print("---")
print(f" 0.3 quantile: {shopping_trends['Purchase Amount (USD)'].agg(pct30)}")
print("---")
print(f"0.4 quantile: {shopping_trends['Purchase Amount (USD)'].agg(pct40)}")

# The .agg() method
```

```python
# print(shopping_trends["'Purchase Amount (USD)'"].cumsum())   # sum
# print(shopping_trends["'Purchase Amount (USD)'"].cummax())   # max 
# print(shopping_trends["'Purchase Amount (USD)'"].cummin())   # min 
# print(shopping_trends["'Purchase Amount (USD)'"].cumprod())  # product
```

```python
# Import NumPy 
import numpy as np

def iqr(column):
    return column.quantile(0.75) - column.quantile(0.25)
    
print(shopping_trends['Purchase Amount (USD)'].agg(iqr))
print("----")
print(shopping_trends[['Purchase Amount (USD)','Previous Purchases']].agg(iqr))
```

```python
print(shopping_trends[['Purchase Amount (USD)','Previous Purchases']].agg([iqr,np.median]))
```
